# Basic usage

```python
# Step 1: Initial settings
%cd /content/
!rm -rf clarinet_plot
!git clone https://github.com/okumakito/clarinet_plot.git
%cd ./clarinet_plot/
import clarinet_plot as cla
```

```python
# Step 2: Load and format data
#   * pandas DataFrame object
#   * rows are observations.
#   * columns are variables such as genes, cell types, time points, experimental conditions, etc.
#   * If each column has different number of observations, set the number of rows to the maximum
#     number of observations, and use None for missing values.
#   * If your data is count data, it is recommended to convert them to log expression data.

# load test data
df = cla.load_test_data()
```

```python
# Step 3: Draw clarinet plots
cla.clarinet_plot(df)
```

# Options

```python
# Change line width (default is 2)
cla.clarinet_plot(df, lw=1)
```

```python
# Change edge color (default is '0.2')
cla.clarinet_plot(df, ec='gray')
```

```python
# Change color map (default is 'husl')
cla.clarinet_plot(df, cmap='tab10')
```

```python
# Change padding from axes (default is 0.05)
cla.clarinet_plot(df, pad=0.2)
```

```python
# Change padding between clarinets (default is 0.1)
cla.clarinet_plot(df, pad_between=1)
```

```python
# Reverse
cla.clarinet_plot(df, ascending=False)
```

```python
# Horizontal
cla.clarinet_plot(df, vertical=False)
```

```python
# Tailless
cla.clarinet_plot(df, show_tail=False)
```

```python
# Use rank instead of quantile
cla.clarinet_plot(df, use_rank=True)
```

```python
# Half
cla.clarinet_plot(df, half=True)
```

```python
# Duet with violin
cla.clarinet_plot(df, duet=True)
```

```python
# Swap duet positions
cla.clarinet_plot(df, duet=True, swap_duet=True)
```

```python
# Set color map for kernel density estimation
cla.clarinet_plot(df, duet=True, cmap_kde='tab10')
```

```python
# Heatmap
cla.clarinet_plot(df, heatmap=True)
```

```python
# Change heatmap color map (default is 'magma')
cla.clarinet_plot(df, heatmap=True, cmap='viridis')
```

```python
# Stripe
cla.clarinet_plot(df, stripe=True)
```

```python
# Change stripe number (default is 10)
cla.clarinet_plot(df, stripe=True, n_stripe=20)
```

```python
# Change stripe line width (default is 1)
cla.clarinet_plot(df, stripe=True, lw_stripe=2)
```

```python
# Change stripe edge color (default is '0.2')
cla.clarinet_plot(df, stripe=True, ec_stripe='white')
```

# Advanced options

```python
# If you would like to modify styles more in details,
# prepare an 'ax' object of matplotlib and pass it to clarinet_plot function as an argument.
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
cla.clarinet_plot(df, ax=ax)
ax.set_title('clarinet plot')
fig.show()
```

```python
# Similarly, the second y-axis for duet mode can be modified by using another 'ax' object.
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax2 = ax.twinx()
cla.clarinet_plot(df, duet=True, ax=ax, ax2=ax2)
ax.set_ylabel('Left side label')
ax2.set_ylabel('Right side label')
fig.show()
```
